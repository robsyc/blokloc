```python
import pandas as pd
import urllib.request

url = "https://data.stad.gent/api/explore/v2.1/catalog/datasets/bloklocaties-gent/exports/csv"
urllib.request.urlretrieve(url, "bloklocaties-gent.csv")
df = pd.read_csv("bloklocaties-gent.csv", sep=";")

df_min = df[['id', 'titel', 'teaser_img_url', 'adres', 'totale_capaciteit', 'gereserveerde_plaatsen', 'label_1', 'openingsuren', 'lees_meer', 'geo_punt']]
df_min.head()
```

```python
# find longest string in column
df_min["lees_meer"].str.len().max()
```

```python
# convert df_min to MySQL table
from sqlalchemy import create_engine

# Credentials to database connection
hostname = "localhost"
dbname   = "blokloc"
uname    = "robbe"
pwd      = "robdbd"

engine = create_engine("mysql+pymysql://{user}:{pw}@{host}/{db}".format(host=hostname, db=dbname, user=uname, pw=pwd))
df_min.to_sql("blokloc", con=engine, if_exists='update', index=False)

# read from MySQL table
df_mysql = pd.read_sql_table('bloklocaties', con=engine)
df_mysql.head()
```
